OPEP:
Criada em 14 de setembro de 1960, a Organização dos Países Exportadores de Petróleo (OPEP) é uma organização intergovernamental, que tem como objetivo a centralização da elaboração das políticas sobre produção e venda do petróleo dos países integrantes.

Atualmente, os países membros da OPEP são: Argélia, Angola, Equador, Irã, Iraque, Kuwait, Líbia, Nigéria, Catar, Arábia Saudita, Emirados Árabes Unidos e Venezuela. A Indonésia suspendeu a sua adesão em janeiro de 2009. A sede da OPEP está localizada em Viena, capital da Áustria.


Perguntas:
    A criação de tal organização impacta na produção e exportação de petróleo do país?
    A adesão à mesma impacta na relevância que um país dá ao ramo de petróleo?
    Considerando um possível aumento na produção e na importância do petróleo recorrente da adesão ao grupo, há alguma consequência social para o país?

Hipóteses:
    Por conta da enorme importancia e influência do grupo na produção e venda de petróleo e visto a grande dependência que o petróleo pode causar em uma economia, cria-se a hipótese de que um país, ao entrar para tal grupo, aumente consideravelmente sua produçao e venda com relação a países que não se encontram no mesmo, criando uma economia muito baseada na venda do óleo e que apresenta enorme desigualdade já que quem enriquece, majoritariamente, são os donos de possos e companhias de extração de petróleo.

```python
#Leitura dos arquivos em Excel
import os
meu_diretorio = "/Users\W10\Documents\Insper\Semestre 2\Ciencia dos dados"
os.chdir(meu_diretorio)

#Planilha da Produção de petróleo de cada país em cada ano (por tonelada):

prodpetrol = pd.read_excel('oil_production_total.xlsx')
prodpetrol=prodpetrol.set_index('country')
paises=['Algeria','Angola','Ecuador','Iran','Iraq','Kuwait','Libya','Nigeria','Qatar','Saudi Arabia','United Arab Emirates','Venezuela']
prodpetrol=prodpetrol.loc[:,[1965, 1980,2000,2015]]

#Definindo os membros da OPEP:

prodpetrol["OPEP"]=0
prodpetrol.loc[paises,'OPEP']=1
prodpetrol.OPEP = prodpetrol.OPEP.astype('category')
prodpetrol.OPEP.cat.categories = ("Não pertence", "Membro")
prodpetrol=prodpetrol.loc[:,['OPEP',1965, 1980,2000,2015]]
prodpetrol.head()
```

Planilha da reservas de petróleo disponíveis para extração em cada país (em toneladas de óleo):

```python
reserva = pd.read_excel('oil_proved_reserves_total.xlsx')
reserva=reserva.set_index('country')
reserva=reserva.loc[:,[1980,2000,2015]]
reserva[1965]='NaN'
reserva["OPEP"]=0
reserva.loc[paises,"OPEP"]=1
reserva.OPEP = reserva.OPEP.astype('category')
reserva.OPEP.cat.categories = ("Não pertence", "Membro")
reserva=reserva[['OPEP',1965,1980,2000,2015]]
reserva.head()
```

Porcentagem dos combustíveis na exportação de mercadorias (%):

```python
txpib = pd.read_excel('tx_val_fuel_zs_un.xlsx')
txpib=txpib.set_index('country')
txpib=txpib.loc[:,[1965, 1980,2000,2015]]
txpib["OPEP"]=0
txpib.loc[paises,"OPEP"]=1
txpib.OPEP = txpib.OPEP.astype('category')
txpib.OPEP.cat.categories = ("Não pertence", "Membro")
txpib=txpib[['OPEP',1965,1980,2000,2015]]
txpib.head(10)
```

Coeficiente de Gini que mostra a desigualdade dos países:

```python
indx_gini=pd.read_excel('inequality_index_gini.xlsx')
indx_gini=indx_gini.set_index('country')
indx_gini=indx_gini.loc[:,[1965,1980,2000,2015]]
indx_gini['OPEP']=0
indx_gini.loc[paises,"OPEP"]=1
indx_gini.OPEP=indx_gini.OPEP.astype('category')
indx_gini.OPEP.cat.categories= ("Nâo pertence", "Membro")
indx_gini=indx_gini[['OPEP',1965,1980,2000,2015]]
indx_gini.head()
```

Com os 4 Dataframes baixados e filtrados, foi criada uma função (def join) que recebe um dos anos escolhidos e retorna um Dataframe com as informações dos outros 4 desenvolvidos previamente.

```python
#Função que retorna um join dos 4 dataframes feitos anteriormente em função do ano:
def join(ano):
    result=pd.DataFrame()
    result['country']=(prodpetrol.index)
    result=result.set_index("country")
    result['OPEP']=0
    result.loc[paises]=1
    result.OPEP = result.OPEP.astype('category')
    result.OPEP.cat.categories = ("Não pertence", "Membro")
    result['Produção de petróleo (toneladas)']=prodpetrol[ano]
    result['Reserva acessível (equivalente em toneladas)']=reserva[ano]
    result['% do combustível na exportação de mercadorias']=txpib[ano]
    result['Coeficiente de Gini']=indx_gini[ano]
    return result
```
