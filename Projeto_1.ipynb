```python
%matplotlib inline
import pandas as pd
import numpy as np
import math as math
from scipy import stats
import matplotlib.pyplot as plt
import os
meu_diretorio = "/Users\W10\Documents\Insper\Semestre 2\Ciencia dos dados\Projeto 1 OPEP"
os.chdir(meu_diretorio)

#Importando e filtrando a planilha da Produção de petróleo de cada país em cada ano (por tonelada):

prodpetrol = pd.read_excel('oil_production_total.xlsx')
prodpetrol=prodpetrol.set_index('country')

#Dividindo os países em 3 grupos: OPEP;
                                 #Gpd, que são os países com produção similar aos da OPEP;
                                 #Restante do mundo;
        
#Definindo os membros da OPEP:
paises=['Algeria','Angola','Ecuador','Iran','Iraq','Kuwait','Libya','Nigeria','Qatar','Saudi Arabia','United Arab Emirates','Venezuela']
prodpetrol["OPEP"]=0
prodpetrol.loc[paises,'OPEP']=1
prodpetrol.OPEP = prodpetrol.OPEP.astype('category')
prodpetrol.OPEP.cat.categories = ("Não pertence", "Membro")

#Definindo os grandes produtores de petróleo que não estão na OPEP (gpd)
gpd=['China', 'Canada', 'Brazil', 'Norway','Kazakhstan','Mexico']

prodpetrol["Grande produtor de petróleo"]=0
prodpetrol.loc[paises,'Grande produtor de petróleo']=1
prodpetrol.loc[ gpd,'Grande produtor de petróleo']=1
prodpetrol["Grande produtor de petróleo"] = prodpetrol["Grande produtor de petróleo"].astype('category')
prodpetrol["Grande produtor de petróleo"].cat.categories = ("Não", "Sim")
```

```python
#importando e filtrando a planilha da taxa do combustível nas mercadorias exportadas:
#Txpib= % da mercadoria exportada
txpib = pd.read_excel('tx_val_fuel_zs_un.xlsx')
txpib=txpib.set_index('country')

#Separando a OPEP e o gpd:
txpib["OPEP"]=0
txpib.loc[paises,"OPEP"]=1
txpib.OPEP = txpib.OPEP.astype('category')
txpib.OPEP.cat.categories = ("Não pertence", "Membro")
txpib["Grande produtor de petróleo"]=0
txpib.loc[paises,'Grande produtor de petróleo']=1
txpib.loc[ gpd,'Grande produtor de petróleo']=1
txpib["Grande produtor de petróleo"] = txpib["Grande produtor de petróleo"].astype('category')
txpib["Grande produtor de petróleo"].cat.categories = ("Não", "Sim")
```

```python
#importando e filtrando a planilha do IDH:

idh=pd.read_excel('hdi_human_development_index.xlsx')
idh=idh.set_index('country')
idh['OPEP']=0
idh.loc[paises,"OPEP"]=1
idh.OPEP=idh.OPEP.astype('category')
idh.OPEP.cat.categories= ("Nâo pertence", "Membro")
idh["Grande produtor de petróleo"]=0
idh.loc[paises,'Grande produtor de petróleo']=1
idh.loc[ gpd,'Grande produtor de petróleo']=1
idh["Grande produtor de petróleo"] = idh["Grande produtor de petróleo"].astype('category')
idh["Grande produtor de petróleo"].cat.categories = ("Não", "Sim")

```

```python
#importando e filtrando a planilha da taxa de tecnologia exportada:
paises1=['Algeria','Ecuador','Iran','Iraq','Kuwait','Nigeria','Qatar','Saudi Arabia','United Arab Emirates','Venezuela']

txtec=pd.read_excel('tx_val_tech_mf_zs.xlsx')
txtec=txtec.set_index('country')
txtec['OPEP']=0
txtec.loc[paises1,"OPEP"]=1
txtec.OPEP=txtec.OPEP.astype('category')
txtec.OPEP.cat.categories= ("Nâo pertence", "Membro")
txtec["Grande produtor de petróleo"]=0
txtec.loc[paises1,'Grande produtor de petróleo']=1
txtec.loc[ gpd,'Grande produtor de petróleo']=1
txtec["Grande produtor de petróleo"] = txtec["Grande produtor de petróleo"].astype('category')
txtec["Grande produtor de petróleo"].cat.categories = ("Não", "Sim")

```

Será feita agora uma função para junção dos dados que recebe um ano e retorna os dados daquele ano

```python
#Função que retorna um join dos dataframes feitos anteriormente em função do ano:
def join(ano):
    result=pd.DataFrame()
    result['country']=(prodpetrol.index)
    result=result.set_index("country")
    result['OPEP']=0
    result.loc[paises]=1
    result.OPEP = result.OPEP.astype('category')
    result.OPEP.cat.categories = ("Não pertence", "Membro")
    result["Grande produtor de petróleo"]=0
    result.loc[paises,'Grande produtor de petróleo']=1
    result.loc[ gpd,'Grande produtor de petróleo']=1
    result["Grande produtor de petróleo"] = prodpetrol["Grande produtor de petróleo"].astype('category')
    result["Grande produtor de petróleo"].cat.categories = ("Não", "Sim")
    result['Produção de petróleo (toneladas)']=prodpetrol[ano]
    result['% do combustível na exportação de mercadorias']=txpib[ano]
    result['IDH']=idh[ano]
    result['% de tecnologia de ponta na exportação de mercadorias']=txtec[ano]
    return result
```

Será plotado agora um gráfico da produção de petróleo pelo tempo para ver se os países da OPEP aumentaram consideravelmente sua produção

```python
#Gráfico do crescimento temporal da produção:
prodopep=prodpetrol.loc[paises,:]
prodoutros=prodpetrol.loc[prodpetrol["Grande produtor de petróleo"]=='Não',:]
prodgpd=prodpetrol.loc[gpd,:]
anos=[]
i=1965
while i < 2017:
    anos.append(i)
    i+=1
plt.plot(anos,prodopep.mean(), label='OPEP')
plt.plot(anos,prodgpd.mean(), label='Gpd')
plt.plot(anos,prodoutros.mean(), label= 'outros')

plt.grid(True)
plt.title('Produção de petróleo ao longo dos anos')
plt.xlabel('Ano')
plt.ylabel('Média da produção (em toneladas)')
plt.legend()
plt.show()

#Crescimento da produção:
cprodgpd=(prodgpd[2014].mean()-prodgpd[1990].mean())/prodgpd[1990].mean()*100
cprodopep=(prodopep[2014].mean()-prodopep[1990].mean())/prodopep[1990].mean()*100
cprodoutros=(prodoutros[2014].mean()-prodoutros[1990].mean())/prodoutros[1990].mean()*100

print('Crescimento da média da produção de grandes produtores de petróleo entre 1990 e 2014:{0}%'.format(cprodgpd))
print('Crescimento da média da produção de membros da OPEP entre 1990 e 2014:{0}%'.format(cprodopep))
print('Crescimento da média da produção de outros entre 1990 e 2014:{0}%'.format(cprodoutros))

```

Análise do gráfico:

O gráfico acima mostra a média da produção de petróleo da OPEP (em azul), a média da produção do Gpd (Grandes produtores de petróleo) (em laranja) e a média da produção do restante dos países disponíveis no dataframe. Assim, pode-se analisar o crescimento temporal da produção de petróleo no mundo. Nota-se que nos anos iniciais, a média da produção dos países membros da OPEP é consideravelmente maior do que o restante, o que indica que, na época da criação do grupo (1960), este era responsável por uma enorme parte da produção mundial, indicando, assim, um dos principais objetivos da criação do grupo: controlar o preço do petróleo no mercado mundial (já que controla grande parte da oferta). 

No entanto, pode-se ver que o maior aumento na produção não aconteceu na OPEP, mas sim em países em desenvolvimento como China, Brasil e México (curva laranja) que descobriram  reservas mais recentemente -63% de crescimento contra 49% da OPEP. Para observar se os países da OPEP possuem uma economia dependente do petróleo, foi escolhida a variável % do petróleo nas mercadorias exportadas, já que a exportação é fundamental para a balança comercial de um país e portanto, essencial para a economia do mesmo:

```python
#Gráfico da relevância do petróleo nas exportações dos países ao longo do tempo:
i=1970
anos=[]
while i < 2017:
    anos.append(i)
    i+=1
txpibopep=txpib.loc[paises,anos]
txpiboutros=txpib.loc[txpib["Grande produtor de petróleo"]=='Não',anos]
txpibgpd=txpib.loc[gpd,anos]

plt.plot(anos,txpibopep.mean(),label='OPEP')
plt.plot(anos,txpibgpd.mean(),label='Gpd')
plt.plot(anos,txpiboutros.mean(),label='Outros')
#plt.plot(anos,txpibopep.median())
#plt.plot(anos,txpiboutros.median())

plt.grid(True)
plt.title('% do combustível nas mercadorias exportadas ao longo dos anos')
plt.xlabel('Ano')
plt.ylabel('% do combustível nas exportações')
plt.legend()
plt.show()
```

O gráfico acima mostra a % do combustível na exportação das mercadorias, ou seja, mostra a relevância do petróleo para a balança comercial e economia do grupo.

O mesmo não acontece com o Gpd nem com o restante dos países, ou seja, os países membros da OPEP são extremamente mais dependentes do petróleo do que o restante do mundo, o que pode ser um sério problema frente a uma possível diminuição na demanda do mesmo. Construção de um gráfico da produção de petróleo versus IDH para possível correlação entre os indicadores:

```python
def reta(dados_x, dados_y, c):    
    a = dados_y.cov(dados_x) / dados_x.var()
    b = dados_y.mean() - a*dados_x.mean()

    print('Ajuste linear entre X3 e X4:')
    print('Coeficiente linear:', b)
    print('Coeficiente angular:', a)
    plt.title('x versus y')
    plt.xlabel('eixo x')
    plt.ylabel('eixo y')

    plt.plot((dados_x.min(), dados_x.max()), (a*dados_x.min()+b, a*dados_x.max()+b), color=c)

    plt.tight_layout()

fig = plt.figure(figsize=(10, 10))
plt.scatter(opep2000['Produção de petróleo (toneladas)'], opep2000['IDH'], c='red', label='OPEP')
reta(opep2000['Produção de petróleo (toneladas)'], opep2000['IDH'], "red")
plt.scatter(outros2000['Produção de petróleo (toneladas)'], outros2000['IDH'], c='blue',label='não OPEP')
reta(outros2000['Produção de petróleo (toneladas)'], outros2000['IDH'], "blue")
plt.title('Produção de petróleo versus IDH em 2000')
plt.xlabel('Produção de petróleo (toneladas)')
plt.legend()
#plt.yscale('log')
plt.ylabel('IDH')
```

No entanto, observa-se que a correlação entre os indicadores para os países da OPEP tem módulo menor do que para o restante dos pontos, o que mostra que apesar de ser positiva, uma produção alta no caso dos membros da OPEP indica um IDH mais baixo do que para um país com mesma produção de fora do bloco, o que mostra a limitação do bloco do desenvolvimento socioeconomico.

```python
def reta(dados_x, dados_y, c):    
    a = dados_y.cov(dados_x) / dados_x.var()
    b = dados_y.mean() - a*dados_x.mean()

    print('Ajuste linear entre X3 e X4:')
    print('Coeficiente linear:', b)
    print('Coeficiente angular:', a)
    plt.title('x versus y')
    plt.xlabel('eixo x')
    plt.ylabel('eixo y')

    plt.plot((dados_x.min(), dados_x.max()), (a*dados_x.min()+b, a*dados_x.max()+b), color=c)

    plt.tight_layout()

fig = plt.figure(figsize=(10, 10))
plt.scatter(opep2000['% do combustível na exportação de mercadorias'], opep2000['% de tecnologia de ponta na exportação de mercadorias'], c='red')
plt.scatter(outros2000['% do combustível na exportação de mercadorias'], outros2000['% de tecnologia de ponta na exportação de mercadorias'], c='blue')
reta(result2000['% do combustível na exportação de mercadorias'], result2000['% de tecnologia de ponta na exportação de mercadorias'], "G")
plt.title('% do combustível na exportação de mercadorias versus % de tecnologia de ponta na exportação de mercadorias em 2000')
plt.xlabel('% do combustível na exportação de mercadorias')
#plt.yscale('log')
plt.ylabel('% de tecnologia de ponta na exportação de mercadorias')
plt.show()
```

Ao tracar-se o gráfico de dispersão entre essas duas variáveis, vê-se claramente uma correlação negativa entre eles, ou seja, quanto maior a dependência, o peso, do petróleo na exportação de mercadorias, menor é o peso das tecnologias de ponta. Essa correlação mostra que uma economia muito focada em exportar comodittys como o petróleo acaba perdendo força em um mercado com maior valor agregado, como evidenciado na hipótese e explicado na teoria da doença holandesa.
